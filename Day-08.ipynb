## POST Requsts 

```python
# Sending Post Requests

import requests

url = 'https://httpbin.org/post'
data = {
    'name': 'John Doe',
    'password': 'secret123'
}

response = requests.post(url, data=data)
print(response.status_code)

if response.status_code == 200:
    print("Post request was successful.")
    print("Response URL:", response.url)
    print(response.text)
else:
    print("Post request failed with status code:", response.status_code)
```

## Basic Authentication

```python
import requests
 
url = 'https://httpbin.org/get'

username = 'user'
password = 'pass'

response = requests.get(url, auth=(username, password))
print(response)


if response.status_code == 200:
    print("Authentication was successful.")
    print(response.url)
    print(response.text)

else:
    print("Authentication failed with status code:", response.status_code)
```

```python
from bs4 import BeautifulSoup

html_doc = """<html><head><title>The Dormouse's story</title></head>
<body>
<p class="title"><b>The Dormouse's story</b></p>
<p class="story">Once upon a time there were three little sisters; and their names were
<a href="http://example.com/elsie&quot; class="sister" id="link1">Elsie</a>,
<a href="http://example.com/lacie&quot; class="sister" id="link2">Lacie</a> and
<a href="http://example.com/tillie&quot; class="sister" id="link3">Tillie</a>;
and they lived at the bottom of a well.</p>
<p class="story">...</p>
"""

soup = BeautifulSoup(html_doc, 'html.parser')  

print(soup.title) # This will print the <title> tag in the HTML document
print(soup.p)  # This will print the first <p> tag in the HTML document
print(soup.a)  # This will print the first <a> tag in the HTML document
p_tag = soup.find('p')
print(p_tag)  # This will print the first <p> tag in the HTML document
p_tags = soup.find_all('p')
print(p_tags)  # This will print all <p> tags in the HTML document
```

## API Integration

```python
import requests
import json
from bs4 import BeautifulSoup

url ='https://api.open-meteo.com/v1/forecast'

params = {
    'latitude': 34.52,
    'longitude': 26.41,
    'current' : 'temperature_2m'
   
}

response = requests.get(url, params=params)
if response.status_code == 200:
    content_type = response.headers.get('Content-Type', '')
    print("Content-Type:", content_type)
    print("successful")
   
    data = response.json()
    print("Weather data:",data)
else:
    print("Request failed with status code:", response.status_code)
```

```python
import requests

url ='https://api.open-meteo.com/v1/forecast'

def fetch_weather(latitude, longitude, current_info):
    params = {
        'latitude': latitude,
        'longitude': longitude,
        'current' : current_info
    }
    response = requests.get(url, params=params)
    
    if response.status_code == 200:
        data = response.json()
        return data
    else:
        return None
    
def parse_weather_data(weather_data):
    if weather_data is None or 'current' not in weather_data:
        print("No weather data available.")
        return None
    current_weather = weather_data['current']['temperature_2m']
    wind_speed = weather_data['current']['windspeed_10m']
    return {
        'temperature': current_weather,
        'wind_speed': wind_speed
    }
  
weather_data = fetch_weather(34.52, 26.41, 'temerature_2m,windspeed_10m')
parsed_data = parse_weather_data(weather_data)
print(parsed_data)


```
